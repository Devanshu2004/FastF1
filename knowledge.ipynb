```python
import fastf1
import matplotlib.pyplot as plt
import pandas as pd
```

```python
session = fastf1.get_session(2019, 'Monza', 'R')
session.load()
```

```python
# Get list of drivers in the particular session (We get the number of driver)
list_of_drivers = session.drivers
print(list_of_drivers)
```

```python
"""
session.results is a table with final data
The columns included are :-
    'DriverNumber',
    'BroadcastName',
    'Abbreviation',
    'DriverId',
    'TeamName',
    'TeamColor',
    'TeamId',
    'FirstName',
    'LastName',
    'FullName',
    'HeadshotUrl',
    'CountryCode',
    'Position',
    'ClassifiedPosition',
    'GridPosition', 
    'Q1',
    'Q2',
    'Q3',
    'Time',
    'Status',
    'Points'
"""
session_results = session.results
df = pd.DataFrame(session_results)
print(df.head())
```

```python
"""
Most important function,
we can get complete analysis for the below given
attributes using 'session.laps' :-
    'Time', 'Driver', 'DriverNumber', 'LapTime', 'LapNumber', 'Stint',
    'PitOutTime', 'PitInTime', 'Sector1Time', 'Sector2Time', 'Sector3Time',
    'Sector1SessionTime', 'Sector2SessionTime', 'Sector3SessionTime',
    'SpeedI1', 'SpeedI2', 'SpeedFL', 'SpeedST', 'IsPersonalBest',
    'Compound', 'TyreLife', 'FreshTyre', 'Team', 'LapStartTime',
    'LapStartDate', 'TrackStatus', 'Position', 'Deleted', 'DeletedReason',
    'FastF1Generated', 'IsAccurate'
"""

laps_data = session.laps
df = pd.DataFrame(laps_data)

print(df.loc[df['Driver'] == 'LEC', ['LapTime', 'LapNumber', 'Position']])
```

```python
# Code to select a specific driver
lap = session.laps.pick_drivers('LEC').pick_fastest()

# Get all the x and y coordinates of the driver during the entire race
x = lap.telemetry['X']
y = lap.telemetry['Y']

# Get speed of the driver for all those coordinates
speed = lap.telemetry['Speed']

# Get gear status of driver
gear = lap.telemetry['nGear']

session_time = lap.telemetry['SessionTime']
rpm = lap.telemetry['RPM']
drs = lap.telemetry['DRS']
brake = lap.telemetry['Brake']

df = pd.DataFrame({
    'x': x,
    'y': y,
    'speed': speed,
    'gear': gear,
    'position': lap['Position'],
    'session time': session_time,
    'rpm': rpm,
    'drs': drs,
    'brake': brake
})

df.head(20)
```
